#### import the main module in matplot lib (pyplot)

```python
from matplotlib import pyplot as plt
#use the magic ntation in ipython notebook to make matplot lib plots in the notebook
%matplotlib inline
```

####Basic Matplotlib 

```python
import numpy as np 
x = np.linspace(0,2,20)
y = x**2
y2 = x**3
# make a figure
fig = plt.figure()
#make your axes (add_axes)
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'b')
axes.plot(x,y2,'r')
#customize (set xlabel ,ylabel, legend)
axes.set_xlabel(r'x $points$', fontsize=14)
axes.set_ylabel(r'y $points$', fontsize=14)
axes.legend([r"$y = x^2 $", r"$y = x^3 $"], loc=2, fontsize=14)
#show your figure
fig.show()
```

# More Figures

```python
xx = np.linspace(-0.75, 1., 100)
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(xx, xx + 0.25*np.random.random_sample(len(xx)))
axes[0].set_title("scatter")

axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5);
axes[3].set_title("fill_between");
```
